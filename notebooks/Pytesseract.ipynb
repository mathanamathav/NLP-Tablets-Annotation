```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import re

req = Request("https://www.netmeds.com/prescriptions/alzheimer")
html_page = urlopen(req)

soup = BeautifulSoup(html_page, "html.parser")

links = []
links = [link.get('href') for link in soup.find_all('a') if re.search(r"tablet|capsule", str(link.get('href')), re.IGNORECASE)]

print(links)
```

```python
import requests
from io import BytesIO
from PIL import Image
```

```python
def scrape_image(url):
    response = Request(url)
    html_page = urlopen(response)
    soup = BeautifulSoup(html_page, 'html.parser')
    images = [img.get('src') for img in soup.find_all('img') if  re.search(r"600x600.*tablet|capsule", str(img.get('src')), re.IGNORECASE)]
    return images

all_images = []

for link in links:
    images = scrape_image(link)
    if images:
        all_images.extend(images)
```

```python
def levenshtein_distance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2+1]
        for i1, c1 in enumerate(s1):
            if c1 == c2:
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1], distances_[-1])))
        distances = distances_
    return distances[-1]
```

```python
def spelling_correction(sentence):
    splittedsentence = word_tokenize(sentence)
    vocwords = list(itertools.chain.from_iterable([vocab.Words.values]))
    for i,word in enumerate(splittedsentence):
        print(word)
        
        if (word not in vocwords and not word.isdigit()): # ignore digits
            levdistances = []
            for vocword in vocwords:
                print(vocword)
                levdistances.append(levenshtein_distance(word,vocword))
            splittedsentence[i] = vocwords[levdistances.index(min(levdistances))]
        else:
            splittedsentence[i] = word
    return ' '.join(splittedsentence)
```
